```python
fig=plt.figure(1)
ax2=fig.add_subplot(111)#1 row,1 coulumn, 1 graph
ax2.set_xlabel('Ticks')
ax2.set_ylabel('Price')
ax2.set_title('Telsa')
ax2.plot('Ticks','Open',data=f)
ax2.plot('Ticks','High',data=f)
ax2.plot('Ticks','Low',data=f)
ax2.plot('Ticks','Close',data=f)
ax2.legend(['Open','High','Low','Close'],prop={'size': 30})
fig.set_size_inches(30,20)
```
